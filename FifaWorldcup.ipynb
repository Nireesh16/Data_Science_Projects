# FIFA World Cup Data Analysis

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Load the dataset
df=pd.read_csv(r"C:\Users\Asus\Desktop\EDAproject\eda project datasets\WorldCups.csv",encoding='utf-8')
df.head()
```

```python
# Check basic information about the dataset
df.info()
```

```python
# Summary statistics
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Convert 'Year' to datetime format
df['Year']=pd.to_datetime(df['Year'],format='%Y')
```

```python
# Visualize the distribution of the 'GoalsScored' variable
sns.histplot(df['GoalsScored'], bins=20, kde=True)
plt.title('Distribution of Goals Scored in FIFA World Cup')
plt.xlabel('Goals Scored')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize the correlation matrix
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Plot the countries with the most wins
top_winners = df['Winner'].value_counts().head(10)
top_winners.plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries with the Most Wins')
plt.xlabel('Country')
plt.ylabel('Number of Wins')
plt.show()
```
