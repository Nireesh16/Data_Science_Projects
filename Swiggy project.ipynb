# "Data Analysis and Preprocessing For Swiggy Dataset"

```python
## Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset
df=pd.read_csv(r"C:\Users\Asus\Desktop\EDAproject\swiggy.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Distribution of average ratings
sns.histplot(df['Avg ratings'],bins=10,kde=True)
plt.title('Distribution of Average Ratings')
plt.xlabel('Average Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation matrix
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Count of restaurants in each city
sns.countplot(x='City',data=df)
plt.title('Count of Restaurants in Each City')
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```

```python
# Convert categorical variables to numerical using one-hot encoding
df = pd.get_dummies(df, columns=['City', 'Food type'], drop_first=True)
```
