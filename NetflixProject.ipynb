# Exploratory Data Analysis (EDA) and Data Preprocessing for Netflix Dataset

```python
#import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df=pd.read_csv(r"C:\Users\Asus\Desktop\EDAproject\netflix_titles.csv")
df.head()
```

```python
# Check basic information about the dataset
df.info()
```

```python
# Summary statistics for numerical columns
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Handle missing values
df['director'].fillna('Unknown',inplace=True)
df['cast'].fillna('Unknown',inplace=True)
df['country'].fillna('Unknown',inplace=True)
df['rating'].fillna('Unknown',inplace=True)
df['duration'].fillna('Unknown',inplace=True)
```

```python
# Convert 'date_added' to datetime format
df['date_added']=pd.to_datetime(df['date_added'])
```

```python
# Extract month and year from 'date_added'
df['added_year']=df['date_added'].dt.year
df['added_month'] = df['date_added'].dt.month
```

```python
# Countplot for 'type'
sns.countplot(x='type', data=df)
plt.title('Distribution of Types (TV Show/Movie)')
plt.show()
```

```python
# Distribution of 'release_year'
sns.histplot(df['release_year'],bins=20,kde=True)
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```

```python
# Top 10 countries with the most content
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='bar')
plt.title('Top 10 Countries with the Most Content')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

# conclusion
This code conducts exploratory data analysis (EDA) and data preprocessing on a Netflix dataset. It loads the data, examines basic information and summary statistics, handles missing values, converts date formats, and visualizes key insights, including the distribution of content types and release years, as well as the top 10 countries with the most content. The process enhances data quality and provides meaningful visualizations for better understanding and interpretation.
